```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
ttf_path = japanize_matplotlib.get_font_ttf_path()
```

```python
df1 = pd.read_csv('../data/app.powerbi.com/2020/生活保護世帯と大学進学率関係.csv')
df1
```

```python
fig, ax = plt.subplots()
ax.scatter(x=df1['生活保護世帯進学率(%)'], y=df1['全世帯進学率(%)'], c='DarkBlue')
for idx, row in df1.iterrows():
    ax.annotate(row['都道府県名'], (row['生活保護世帯進学率(%)'], row['全世帯進学率(%)']))
plt.show()
```

```python
df2 = pd.read_excel('../data/www.mhlw.go.jp/2020/都道府県別被保護者調査.xls')
df2 = df2.shift(-5)[0:47]
columns = ['無１', '都道府県', '無２', '被保護世帯総数', '現に保護を受けた世帯', '保護停止中の世帯', '被保護実人員総数', '現に保護を受けた人数', '保護停止中の人数', '保護開始世帯数', '保護廃止世帯数']
df2.columns = columns
df2
```

```python
df_concat1 = pd.concat([df1, df2], axis=1)
df_concat1
```

```python
fig, ax = plt.subplots()
ax.scatter(x=df_concat1['生活保護世帯進学率(%)'], y=df_concat1['被保護世帯総数'], c='DarkBlue')
for idx, row in df_concat1.iterrows():
    ax.annotate(row['都道府県名'], (row['生活保護世帯進学率(%)'], row['被保護世帯総数']))
plt.show()
```

```python
df3 = pd.read_excel('../data/www.soumu.go.jp/2020/都道府県別人口と世帯数.xls')
df3 = df3.shift(-4)[0:47]
columns = [
  '団体コード', '都道府県名', '人口 男', '人口 女', '人口 男女合計', '世帯数',
  'Skip A', 'Skip B', 'Skip C', 'Skip D', 'Skip E', 'Skip F', 'Skip G',
  'Skip H', 'Skip I', 'Skip J', 'Skip K', 'Skip L', 'Skip M', 'Skip N',
  'Skip O', 'Skip P', 'Skip Q', 'Skip R'
  ]
df3.columns = columns
df3
```

```python
df_concat2 = pd.concat([df_concat1, df3], axis=1)
df_concat2['生活保護世帯比率'] = df_concat2['現に保護を受けた世帯'] / df_concat2['世帯数']
df_concat2
```

```python
fig, ax = plt.subplots()
ax.scatter(x=df_concat2['生活保護世帯進学率(%)'], y=df_concat2['生活保護世帯比率'], c='DarkBlue')
for idx, row in df_concat2.iterrows():
    ax.annotate(row['都道府県'], (row['生活保護世帯進学率(%)'], row['生活保護世帯比率']))
plt.show()
```
